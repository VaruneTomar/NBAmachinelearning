```python
import pandas as pd
```

```python
box_scores = pd.read_csv("nbascoresdata.csv", index_col=0)
box_scores
```

```python
box_scores = box_scores.sort_values("date")
```

```python
box_scores = box_scores.reset_index(drop=True)
```

```python
def nextGame(group):
    group["next_game"] = group["won"].shift(-1)
    return group

box_scores = box_scores.groupby("team", group_keys=False).apply(nextGame)
```

```python
box_scores["next_game"][pd.isnull(box_scores["next_game"])] = 2
box_scores["next_game"] = box_scores["next_game"].astype(int, errors="ignore")
```

```python
nulls = pd.isnull(box_scores).sum()
nulls = nulls[nulls > 0]
valid_columns = box_scores.columns[~box_scores.columns.isin(nulls.index)]
box_scores = box_scores[valid_columns].copy()
```

```python
box_scores
```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = RidgeClassifier(alpha=1)

split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split, n_jobs=1)
```

```python
removed_columns = ["season", "date", "won", "next_game", "team", "team_opp"]
selected_columns = box_scores.columns[~box_scores.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
box_scores[selected_columns] = scaler.fit_transform(box_scores[selected_columns])
```

```python
box_scores
```

```python
sfs.fit(box_scores[selected_columns], box_scores["next_game"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
```

```python
predictors
```

```python
def backtest(scores, datamodel, predictors, start=3, step=1):
    all_predictions = []
    
    seasons = sorted(scores["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i]
        train = scores[scores["season"] < season]
        test = scores[scores["season"] == season]
        
        datamodel.fit(train[predictors], train["next_game"])
        
        preds = datamodel.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["next_game"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(box_scores, rr, predictors)

from sklearn.metrics import accuracy_score

accuracy_score(predictions["actual"], predictions["prediction"])
```
